```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Load your dataset
data = pd.read_csv('MOCK_DATA_Bangalore.csv')
area_column = data['Area']
sl_no = data['sl']
# Drop the first column
data.drop(columns=['sl','Area'], inplace=True)

# Preprocess the data (scale the features)
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

# Perform dimensionality reduction using PCA
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(scaled_data)

# Choose the number of clusters (K)
k = 3  # Choose an appropriate value for K

# Apply K-means clustering
kmeans = KMeans(n_clusters=k)
kmeans.fit(scaled_data)
labels = kmeans.labels_

# Add cluster labels to the original dataset
data['Cluster'] = labels
data.insert(0, 'Area', area_column)
data.insert(0, 'Sl No.', sl_no)

# Optional: Evaluate the clustering
# For example, you can use silhouette score
from sklearn.metrics import silhouette_score
silhouette_score = silhouette_score(scaled_data, labels)
print("Silhouette Score:", silhouette_score)

# Visualize the clusters after dimensionality reduction
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=labels, cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('K-means Clustering after Dimensionality Reduction')
plt.show()

# Save the clustered dataset
data.to_csv('clustered_data.csv', index=False)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from sklearn.preprocessing import StandardScaler, LabelEncoder

# Load your dataset
data = pd.read_csv('MOCK_DATA_Bangalore.csv')

# Preprocess the data (scale the features)
area_column = data['Area']
sl_no = data['sl']
# Drop the first column
data.drop(columns=['sl','Area'], inplace=True)

scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

# Apply Gaussian Mixture Model clustering
k = 3  # Number of clusters (surplus, scarcity, normal)
gmm = GaussianMixture(n_components=k, random_state=42)
gmm.fit(scaled_data)
labels = gmm.predict(scaled_data)

# Add cluster labels to the original dataset
data['Cluster'] = labels
data.insert(0, 'Area', area_column)
data.insert(0, 'Sl No.', sl_no)

# Create a heatmap of cluster probabilities
cluster_probabilities = gmm.predict_proba(scaled_data)
plt.figure(figsize=(10, 8))
sns.heatmap(cluster_probabilities, cmap='Blues', cbar=True, xticklabels=['Surplus', 'Scarcity', 'Normal'])
plt.xlabel('Cluster')
plt.ylabel('Data Point')
plt.title('Cluster Probabilities')
plt.show()

from sklearn.metrics import silhouette_score
silhouette_score = silhouette_score(scaled_data, labels)
print("Silhouette Score:", silhouette_score)

# Visualize the clusters after dimensionality reduction
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=labels, cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')

plt.title('GMM Clustering after Dimensionality Reduction')
plt.show()


# Save the clustered dataset
data.to_csv('clustered_data_GMM.csv', index=False)

```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Load your dataset
data = pd.read_csv('MOCK_DATA_Bangalore.csv')

# Preprocess the data (scale the features)
area_column = data['Area']
sl_no = data['sl']
# Drop the first column
data.drop(columns=['sl','Area'], inplace=True)

scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

# Apply PCA for dimensionality reduction
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(scaled_data)

# Apply DBSCAN clustering
dbscan = DBSCAN(eps=0.5, min_samples=5)
labels = dbscan.fit_predict(reduced_data)
data['Cluster'] = labels
data.insert(0, 'Area', area_column)
data.insert(0, 'Sl No.', sl_no)

from sklearn.metrics import silhouette_score
silhouette_score = silhouette_score(scaled_data, labels)
print("Silhouette Score:", silhouette_score)

# Visualize the clustering
plt.figure(figsize=(10, 8))
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=labels, cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('DBSCAN Clustering')
plt.colorbar(label='Cluster')
plt.show()

# Save the clustered dataset
data.to_csv('clustered_data_DBSCAN.csv', index=False)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder, RobustScaler
from sklearn.decomposition import PCA
from sklearn.cluster import Birch

# DATA
data = pd.read_csv('MOCK_DATA_Bangalore.csv')

area_column = data['Area']
sl_no = data['sl']

# Drop the first column
data.drop(columns=['sl','Area'], inplace=True)

scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

# Apply PCA for dimensionality reduction
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(scaled_data)

# Creating the BIRCH clustering model
birch = Birch(branching_factor = 50, n_clusters = 3, threshold = 1.5)
labels = birch.fit_predict(reduced_data)
data['Cluster'] = labels
data.insert(0, 'Area', area_column)
data.insert(0, 'Sl No.', sl_no)

# Visualize the clustering
plt.figure(figsize=(6, 4))
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=labels, cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('BIRCH Clustering')
plt.colorbar(label='Cluster')
plt.show()

# Save the clustered dataset
data.to_csv('clustered_data_BIRCH.csv', index=False)
```

```python
import random

class AntColony:
    def __init__(self, num_ants, num_iterations, pheromone_evaporation_rate, alpha, beta):
        self.num_ants = num_ants
        self.num_iterations = num_iterations
        self.pheromone_evaporation_rate = pheromone_evaporation_rate
        self.alpha = alpha
        self.beta = beta

    def optimize(self, graph):
        best_path = None
        best_distance = float('inf')

        num_nodes = len(graph)
        pheromone = [[1 / (num_nodes * num_nodes) for _ in range(num_nodes)] for _ in range(num_nodes)]

        for _ in range(self.num_iterations):
            paths = []
            for _ in range(self.num_ants):
                path = self.generate_path(graph, pheromone)
                paths.append(path)
                distance = self.calculate_distance(graph, path)
                if distance < best_distance:
                    best_path = path
                    best_distance = distance

            self.update_pheromone(graph, pheromone, paths)

        return best_path, best_distance

    def generate_path(self, graph, pheromone):
        num_nodes = len(graph)
        start_node = random.randint(0, num_nodes - 1)
        visited = [False] * num_nodes
        visited[start_node] = True
        path = [start_node]

        while len(path) < num_nodes:
            current_node = path[-1]
            probabilities = self.calculate_probabilities(graph, pheromone, visited, current_node)
            next_node = self.select_next_node(probabilities)
            path.append(next_node)
            visited[next_node] = True

        return path

    def calculate_probabilities(self, graph, pheromone, visited, current_node):
        probabilities = []
        pheromone_sum = 0

        for node, distance in enumerate(graph[current_node]):
            if not visited[node]:
                pheromone_sum += pheromone[current_node][node] ** self.alpha * (1 / distance) ** self.beta

        for node, distance in enumerate(graph[current_node]):
            if visited[node]:
                probabilities.append(0)
            else:
                probability = pheromone[current_node][node] ** self.alpha * (1 / distance) ** self.beta / pheromone_sum
                probabilities.append(probability)

        return probabilities

    def select_next_node(self, probabilities):
        rand = random.random()
        total = 0
        for i, probability in enumerate(probabilities):
            total += probability
            if rand <= total:
                return i

    def calculate_distance(self, graph, path):
        distance = 0
        for i in range(len(path) - 1):
            distance += graph[path[i]][path[i + 1]]
        return distance

    def update_pheromone(self, graph, pheromone, paths):
        evaporation = 1 - self.pheromone_evaporation_rate
        for i in range(len(pheromone)):
            for j in range(len(pheromone[i])):
                pheromone[i][j] *= evaporation

        for path in paths:
            distance = self.calculate_distance(graph, path)
            for i in range(len(path) - 1):
                pheromone[path[i]][path[i + 1]] += 1 / distance
                pheromone[path[i + 1]][path[i]] += 1 / distance


# Example usage:
if __name__ == "__main__":
    # Example graph - distance between nodes
    graph = [
        [0, 2, 4, 5],
        [2, 0, 3, 2],
        [4, 3, 0, 3],
        [5, 2, 3, 0]
    ]

    num_ants = 10
    num_iterations = 100
    pheromone_evaporation_rate = 0.1
    alpha = 1
    beta = 2

    ant_colony = AntColony(num_ants, num_iterations, pheromone_evaporation_rate, alpha, beta)
    best_path, best_distance = ant_colony.optimize(graph)
    print("Best Path:", best_path)
    print("Best Distance:", best_distance)
```

```python
from geopy.geocoders import Nominatim
import pandas as pd
import time

geolocator=Nominatim(user_agent="Cache")

df = pd.read_csv('MOCK_DATA_Bangalore.csv')
latitude=[]
longitude=[]

for i in df.iloc[:, 1]:
    try:
        location=geolocator.geocode(i)
        latitude.append(location.latitude)
        longitude.append(location.longitude)
        print(i,location.latitude,location.longitude)
        
    except:
        latitude.append("fail")
        longitude.append("fail")
        print(i,"Failed")
    time.sleep(1)
```
